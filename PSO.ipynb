```python
import numpy as np
```

>here we have a class of particle which contains position and the best position of each particle

```python
class particle:
    def __init__(self,x,y):
        self.pos = np.array([x, y])
        self.best_pos = self.pos
```

>we define a fitness function, which is the function from the given problem

```python
def fitness_func(arr):
    x = arr[0]
    y = arr[1]
    return abs(np.sin(x)*np.cos(y)*np.exp(abs(1 - (np.sqrt(x**2 + y**2)/np.pi))))
```

>this is a check function which project the points in [-10, 10]Ã—[-10, 10]

```python
def check(arr):
    if (arr[0] < -10): arr[0] = -10
    if (arr[0] > 10): arr[0] = 10
    if (arr[1] < -10): arr[0] = -10
    if (arr[1] > 10): arr[0] = 10
```

this is the main function:
>first in part A we define particles and calculate their fitnesses

>in part B, we define a globla best particle

>part C determines each generation

>in part D, we move each particle be the formula:
$$ x_{new} = x + c_1*rand*(G_b - x) + c_2*rand*(x_b - x) $$
>then in part E, we check for the global best update

```python
def PSO(n, iterations, c_1, c_2, w):
    particles = [particle(np.random.uniform(-10, 10), np.random.uniform(-10, 10)) for i in range(n)] #part A
    fitnesses = [fitness_func(particles[i].pos) for i in range(n)]
    
    maximum = 0
    max_i = 0
    for i in range(n):
        if (maximum < fitnesses[i]):
            maximum = fitnesses[i]
            max_i = i
    
    best_global_index = i #part B
    best_fitness = maximum
    
    best = []
    
    for it in range(iterations): #part C
        for i in range(n): #part D
            particles[i].pos = (particles[i].pos) + c_1*np.random.rand(1)*(particles[best_global_index].pos - particles[i].pos) + c_2*np.random.rand(1)*(particles[i].best_pos - particles[i].pos)
            
            check(particles[i].pos) #part E
            if (fitness_func(particles[i].pos) > fitness_func(particles[i].best_pos)):
                particles[i].best_pos = particles[i].pos
                
            if (fitness_func(particles[i].pos) > fitness_func(particles[best_global_index].best_pos)):
                best_global_index = i
                
        best.append(fitness_func(particles[best_global_index].best_pos))
    return best
```

```python
import matplotlib.pyplot as plt
```

```python
arr = PSO(100, 50, 0.6, 0.7, 0.7)
ite = [i+1 for i in range(50)]
plt.plot(ite, arr)
plt.show()
arr
```
